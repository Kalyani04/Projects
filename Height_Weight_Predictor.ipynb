```python
#MSE( cost ) = (Sum ( y - (m * x + b) )**2 ) / N = sum((y_actual - y_predicted)*
#dM/dm = 2/n (sum( -x (y -( mx + b))))
#dM/db = 2/n (sum (-(y - (mx + b))))
#m_current = m_current -learning_rate * dM/dm
#b_current = b_current - learning_rate *dM/db
#x = height
#y = weight
def gradient_descent( x, y, iterations, learning_rate ):
 m_current = 0
 b_current = 0
 N = len(x)
# learning_rate = 0.1


 for i in range(iterations):
    y_predicted = [None] * N
    Cost = 0
    m_derivative = 0
    b_derivative = 0
 #  print(y_predicted)
    
    for index in range(len(x)):
        y_predicted[index] = m_current * x[index][0] + b_current
        #print(y_predicted)
        Cost = Cost + ((weight[index] - y_predicted[index]) **2)
        #print(Cost)
        m_derivative = m_derivative + ( -1 * x[index][0] * ( weight[index] - y_predicted[index] ))
       #print(m_derivative)
        b_derivative = b_derivative * ( -(weight[index] - y_predicted[index] ))
        Cost = Cost / N
        m_derivative = m_derivative * (2 / N) 
        b_derivative = b_derivative * (2 / N)
                                          
        m_current = m_current - (learning_rate * m_derivative)
        b_current = b_current -(learning_rate * b_derivative)
       #print(y_predicted)
        pd.plot(height, y_predicted, color = 'b')
        pd.scatter(b_current, Cost, color ='navy')
       #pd.scatter(height, weight, color='g')


       #pd.show()
    print(" m_current : {} , b_current : {}, Cost : {} , iteration : {} ".format(  m_current, b_current, Cost, i ))
  #  return m_current, b_current
value = gradient_descent(height, weight, 20, 0.001 )
#print(value)
pd.show()
```
